```python
# A Python program to generate word cloud of lyrics from your favourite song

# Research pointers
# I am working with Text Data (Lyrics of a song to be particular)
# I am using word cloud to plot most commonly used words in the song

# Import Libraries NLTK, MatPlot and WordCloud
import nltk
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# This is a function that accepts a filename and fileoption
# It then generates a list of words in song
# Close the file as we will not be needing it post this function
def generateWords(filename, fileOption):
    word = []
    file = open(filename, fileOption)
    for line in file:
        if line.strip() != ' ':
            word.extend(line.strip().split())
    return word
    file.close()

# This function accepts the list of words and converts it to a string
# We use returned string to generate the word cloud
def generateStringFromLines(word):
    string = ""
    for i, w in enumerate(word):
        #Skip adding space if first item in list
        if i == 0:
            string += w
        else:    
            string += " " +w
    return string

# This function generates the word cloud from the string
# I Then use plt library to plot the wordcloud
def generateWorldCloud(cloudString):
    wordcloud = WordCloud().generate(cloudString)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()

# These lines call all the functions with right function parameters to generate the word cloud
# You can make a word cloud of a song you like, simply by putting the lyrics in the lyrics.txt file in the same folder
# I have chosen Bob Dylan's "Like A Rolling Stone"
songWords = generateWords("lyrics.txt", "r")           
songString = generateStringFromLines(songWords)
generateWorldCloud(songString)
```
